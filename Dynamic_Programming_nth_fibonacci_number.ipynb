```python
'''

Write a function fib(n) that takes in a number as in argument. The fucntion should returen the n-th number of the fibonacci sequence.


nth number in the fibonacci sequence is n-1  +  n-2 


'''
```

```python
def fib(n):

    if n <= 2:
        return 1

    return fib(n-1) + fib(n-2)

#works for smaller numbers like 6, 7, 8 but for bigger numbers like 50 it is stuck or taking too long 

#take a look below multiple numbers are being recalculated making it inefficient for bigger numbers 

#the function is 2^n time complexity
```

```python
#efficient method

#implement memoization - derived from using a memo where you store a certain information 

# using a dictionary to store the values


def fib(n , memo = {}):

    if n in memo:
        return memo[n]

    if n <= 2:
        return 1

    memo[n] = fib(n-1,memo) + fib(n-2,memo)

    return memo[n]



```

```python
print(fib(50))
```
