```python
#writing a recursive fucntion for it 

def canSum ( targetSum, numbers ):

    if targetSum == 0 : return True

    if targetSum < 0 : return False

    #now for each number in the array we can write a recursive call

    for num in numbers:
        remainder = targetSum - num
        if canSum(remainder,numbers):
            return True
    return False
```

```python
print(canSum(7,[2,3])) #won't work for larger target sums 
```

```python
def canSumNew(targetSum, numbers, memo={}):

    if targetSum in memo:
        return memo[targetSum]

    if targetSum == 0 : return True

    if targetSum < 0 : return False

    #now for each number in the array we can write a recursive call

    for num in numbers:
        remainder = targetSum - num

        memo[remainder] = canSumNew(remainder,numbers,memo)

        if memo[remainder]:
            return True
    return False
```

```python
print(canSumNew(300,[2,3,5]))
```
