```python

def gridTraveller(m,n, memo={}):


    #base cases

    if m==1 and n==1 : return 1

    if m==0 or n==0 : return 0


    #when going down the number of rows decrease by 1 and when going right the number of columns decrease by 1 

    return gridTraveller(m-1, n) + gridTraveller(m,n-1)


```

```python
print(gridTraveller(3,3)) # but won't work for 18,18
```

```python

def gridTravellerNew(m,n, memo={}):


    if (m,n) in memo:
        return memo[(m,n)]
    if (n,m) in memo:
        return memo[(n,m)]

    #base cases

    if m==1 and n==1 : return 1

    if m==0 or n==0 : return 0


    #when going down the number of rows decrease by 1 and when going right the number of columns decrease by 1 

    memo[(m,n)] = memo[(n,m)] =  gridTravellerNew(m-1, n,memo) + gridTravellerNew(m,n-1,memo)

    return  memo[(m,n)]


```

```python
print(gridTravellerNew(18,18))
```
